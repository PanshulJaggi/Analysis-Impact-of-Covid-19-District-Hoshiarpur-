# Covid-19 ( INDIA )

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np

# loading the Covid-19 (India) DataSet
data = pd.read_csv("C:\\Users\\pansh\\Downloads\\covid_19_india.csv", parse_dates=['Date'], dayfirst=True)
data
```

```python
data.shape
```

```python
data.info()
```

```python
# keeping only required columns
```

```python
data = data[ ['Date','State/UnionTerritory','Cured','Deaths','Confirmed'] ]
data
```

```python
# Check the number of missing values in each column.
data.isnull().sum()
```

```python
# renaming the column name
```

```python
data.columns = ['Date','State','Cured','Deaths','Confirmed']
data
```

```python
# looking at the latest dates
```

```python
data.tail()
```

```python
# current date
```

```python
today = data[ data.Date == '2021-08-11' ]
today.head()
```

```python
# sorting data w.r.t number of the confirmed cases
```

```python
maximum_confirmed_cases = today.sort_values(by='Confirmed', ascending=False)
maximum_confirmed_cases
```

```python
# getting top 5 states with maximun number of confirmed cases in today
```

```python
top_states_confirmed_cases = maximum_confirmed_cases[0:5]
top_states_confirmed_cases
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State',y='Confirmed', data=top_states_confirmed_cases, hue='State')

plt.title('Top 5 states with maximun number of confirmed cases in today', fontsize=14)
plt.xlabel('States', fontsize=16)
plt.ylabel('No. of Confirmed Cases', fontsize=16)

plt.show()
```

```python
# sorting data w.r.t number of the deaths cases
```

```python
maximum_deaths_cases = today.sort_values(by='Deaths', ascending=False)
maximum_deaths_cases
```

```python
# getting top 5 states with maximun number of deaths cases in today
```

```python
top_states_deaths_cases = maximum_deaths_cases[0:5]
top_states_deaths_cases
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State',y='Deaths', data=top_states_deaths_cases, hue='State')

plt.title('Top 5 states with maximun number of deaths cases in today', fontsize=14)
plt.xlabel('States', fontsize=16)
plt.ylabel('No. of Confirmed Cases', fontsize=16)

plt.show()
```

```python
# sorting data w.r.t number of the cured cases
```

```python
maximum_cured_cases = today.sort_values(by='Cured', ascending=False)
maximum_cured_cases
```

```python
# getting top 5 states with maximun number of cured cases in today
```

```python
top_states_cured_cases = maximum_cured_cases[0:5]
top_states_cured_cases
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State',y='Cured', data=top_states_cured_cases, hue='State')

plt.title('Top 5 states with maximun number of cured cases in today', fontsize=14)
plt.xlabel('States', fontsize=16)
plt.ylabel('No. of Confirmed Cases', fontsize=16)

plt.show()
```

```python
# Covid-19 information related to the state of Punjab in India.
```

```python
punjab = data[ data.State == 'Punjab' ]
punjab
```

```python
#  the total number of Confirmed Cases due to COVID-19 in Punjab.
punjab.Confirmed.sum()
```

```python
#  the total number of Deaths Cases due to COVID-19 in Punjab.
punjab.Deaths.sum()
```

```python
# Find the date of the highest deaths in Punjab due to COVID-19 in Punjab.
punjab.loc[punjab['Deaths'].idxmax(),'Date']
```

```python
# Visualizing the trend of confirmed COVID-19 cases over time in the state of Punjab.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x='Date',y='Confirmed', data=punjab, color='g')

plt.title(' Display the trend of confirmed COVID-19 cases over time in the state of Punjab', fontsize=14)

plt.show()
```

```python
# Visualizing the trend of deaths COVID-19 cases over time in the state of Punjab.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x='Date',y='Deaths', data=punjab, color='r')

plt.title(' Display the trend of deaths COVID-19 cases over time in the state of Punjab', fontsize=14)

plt.show()
```
